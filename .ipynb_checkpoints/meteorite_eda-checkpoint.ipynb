## Checking that all Latitude and Longitude value are Floats

```python
meteorites_len = meteorites_df.shape[0]
ufo_len = ufo_sightings_df.shape[0]
wonder_len = wonders_df.shape[0]
```

```python
meteorites_df.dtypes['reclat'],meteorites_df.dtypes['reclong']
```

```python
wonders_df.dtypes['Latitude'],wonders_df.dtypes['Longitude']
```

```python
ufo_sightings_df.dtypes['latitude'],ufo_sightings_df.dtypes['longitude']
```

```python
#Convert lat and lon in `ufo_sightings_df` to float

                                #there was an instance where where tehre was a letter in the lat value
ufo_sightings_df['latitude'] = [float('33.200088')
                                if ufo_sightings_df.latitude[i] == '33q.200088'
                                else float(ufo_sightings_df.latitude[i]) for i in np.arange(ufo_len)]
ufo_sightings_df['longitude'] = [float(ufo_sightings_df.longitude[i]) for i in np.arange(ufo_len)]

```

```python
# fill nan value in mass column with 0
meteorites_df.fillna(0,inplace=True)
```

```python
# convert mass column from grams to lbs
meteorites_df['mass'] = [round(meteorites_df.mass[i]/453.59237,2) for i in np.arange(meteorites_len)]
```

```python
# filter out records that have irregular latitude and longitude values, and obtain meteorites that are between 860 BCE and 2016

meteorites_df = meteorites_df.loc[(meteorites_df.year>=860) & (meteorites_df.year<=2016)] # filter out weird years
meteorites_df = meteorites_df.loc[((meteorites_df.reclong<=180) & (meteorites_df.reclong>=-180)) & ((meteorites_df.reclat!=0) | (meteorites_df.reclong!=0))]# filter out weird locations

```

```python
# Adds geomerty column to the dataframes using the lat and lon values
meteorites_df['geometry'] = [Point(xy) for xy in zip(meteorites_df.reclong,meteorites_df.reclat)]
ufo_sightings_df['geometry'] = [Point(xy) for xy in zip(ufo_sightings_df.longitude,ufo_sightings_df.latitude)]
wonders_df['geometry']  = [Point(xy) for xy in zip (wonders_df.Longitude,wonders_df.Latitude)]
```

```python
# add year column to ufo_sightings_df

ufo_sightings_df['year'] = [int(re.findall('\d{4}',date)[0]) for date in ufo_sightings_df['Date_time']]
```

### Working Dataframes

### Density of Phenomena


```python
def phenomena_percent(continent):
    """
    string -> dict
    expects a continent
    returns the percent of all phenomena for the given continent
    
    """
    # locate the data points that fall in the continent with a spatial join
    continent_meteorites_df = gpd.sjoin(geo_meteorites,world_bound.loc[world_bound['continent'] == continent])
    continent_ufo_df = gpd.sjoin(geo_ufo,world_bound.loc[world_bound['continent'] == continent])
    continent_wonder_df = gpd.sjoin(geo_wonder,world_bound.loc[world_bound['continent'] == continent])
    
    # obtain the percantage of each and every phenomena
    num_meteorites = len(continent_meteorites_df)
    num_ufo = len(continent_ufo_df)
    num_wonder = len(continent_wonder_df)
    phenomenon_count =  num_meteorites + num_ufo + num_wonder
    meteorite_percent = round(num_meteorites/total_meteorites* 100,2)
    ufo_percent = round(num_ufo/total_ufo* 100,2)
    wonder_percent = round(num_wonder/total_wonders* 100,2)
    phenomenon_percent = round(phenomenon_count/total_phenomenon* 100,2)
    
    # put them in a dict
    percent_dict = {'total_phenomenon_percent':phenomenon_percent,
                   'meteorite_percent':meteorite_percent,
                   'ufo_percent':ufo_percent,
                   'wonder_percent':wonder_percent}
    return percent_dict
    
    
```

```python
# make a dataframe with all continents and their respective phenomena percentages
# and geometry

phenomena_percent_df = pd.DataFrame({'continent':world_continents})
phenomena_percent_df['total_phenomenon_percent'] = [phenomena_percent(continent)['total_phenomenon_percent'] for continent in world_continents]
phenomena_percent_df['meteorite_percent'] = [phenomena_percent(continent)['meteorite_percent'] for continent in world_continents]
phenomena_percent_df['ufo_percent'] = [phenomena_percent(continent)['ufo_percent'] for continent in world_continents]
phenomena_percent_df['wonder_percent'] = [phenomena_percent(continent)['wonder_percent'] for continent in world_continents]
phenomena_percent_df = gpd.GeoDataFrame(phenomena_percent_df.merge(world_bound,on='continent'))


phenomena_percent_df.head()
```

```python
def phenomena_density(meteorite_percent=False,ufo_percent=False,wonder_percent=False):
    """
    meteorite_percent=boolean ufo_percent=boolean wonder_percent=boolean -> plot
    
    By default, returns a density map of all phenomena.
    If any of the optional arguments, which correspond to a phenomenon, are set to true,
    it will return a density map for that specific phenomena.
    
    """
    if meteorite_percent:

        fig, ax = plt.subplots(figsize=(10,10))

        dens_plot = phenomena_percent_df.plot(ax=ax,cmap='cool_r',edgecolor='k',column='meteorite_percent',
                                 legend=True,legend_kwds={'label':'Phenomena Percent','shrink':0.5,'pad':0.02})

        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

        plt.title('Global Distribution of Meteorites')
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.show()
        
    elif ufo_percent:

        fig, ax = plt.subplots(figsize=(10,10))

        dens_plot = phenomena_percent_df.plot(ax=ax,cmap='cool_r',edgecolor='k',column='ufo_percent',
                                 legend=True,legend_kwds={'label':'Phenomena Percent','shrink':0.5,'pad':0.02})

        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

        plt.title('Global Distribution of UFO Sightings')
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.show()
    elif wonder_percent:

        fig, ax = plt.subplots(figsize=(10,10))

        dens_plot = phenomena_percent_df.plot(ax=ax,cmap='cool_r',edgecolor='k',column='wonder_percent',
                                 legend=True,legend_kwds={'label':'Phenomena Percent','shrink':0.5,'pad':0.02})

        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

        plt.title('Global Distribution of World Wonders')
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.show()
    else:
        
        fig, ax = plt.subplots(figsize=(10,10))

        dens_plot = phenomena_percent_df.plot(ax=ax,cmap='cool_r',edgecolor='k',column='total_phenomenon_percent',
                                 legend=True,legend_kwds={'label':'Phenomena Percent','shrink':0.5,'pad':0.02})

        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)

        plt.title('Global Distribution of Phenomena \n (Meteorites,UFO Sightings,World Wonders)')
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.show()
```

### Examining Continental Phenomena 

```python
def continent_phenomena(continent,meteorite=False,ufo=False,wonders=False):
    """
    string meteorite=boolean ufo=boolean wonders=boolean  -> plot
    expects a name of a continent
    returns a projection of the given continent
    if one the data parameters is set to true
    it will plot the given Phenomena/Phenomenon data
    
    """
     # oceania [xmin, xmax, ymin, ymax]
    oceania_proj =  [95, 180, -60, 13]
    # europe [xmin, xmax, ymin, ymax]
    eu_proj = [-10, 50, 35, 71]
    
    # outputs the continent(all phenomena)
    
    if meteorite and ufo and wonders:  
        
        fig,ax = plt.subplots(figsize=(10,10))
        #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        continent_meteorites_df = gpd.sjoin(geo_meteorites,world_bound.loc[world_bound['continent'] == continent])
        continent_ufo_df = gpd.sjoin(geo_ufo,world_bound.loc[world_bound['continent'] == continent])
        continent_wonder_df = gpd.sjoin(geo_wonder,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        
        
        
        continent_meteorites_df.plot(ax=ax,color='orangered',markersize=15,label='metorites')
        continent_ufo_df.plot(ax=ax,color='lime',markersize=15,marker='v',label='ufo_sightings')
        continent_wonder_df.plot(ax=ax,color='gold',markersize=15,marker='s',label='world_wonder')
        
        
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of all phenomena in the continent
        phenomenon_count = len(continent_ufo_df) + len(continent_meteorites_df) + len(continent_wonder_df)
        phenomenon_percent = phenomena_percent(continent)['total_phenomenon_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12) 
        
        
        fig.show()
        
        
    # outputs the continent( meteorite and ufo data)    
    elif meteorite and ufo:  
        
        fig,ax = plt.subplots(figsize=(10,10))
        #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        continent_meteorites_df = gpd.sjoin(geo_meteorites,world_bound.loc[world_bound['continent'] == continent])
        continent_ufo_df = gpd.sjoin(geo_ufo,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        
        
        
        continent_meteorites_df.plot(ax=ax,color='orangered',markersize=15,label='metorites')
        continent_ufo_df.plot(ax=ax,color='lime',markersize=15,marker='v',label='ufo_sightings')
        
        
        
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of all phenomena in the continent
        phenomenon_count = len(continent_ufo_df) + len(continent_meteorites_df)
        phenomenon_percent = phenomena_percent(continent)['total_phenomenon_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12) 
        
        
        fig.show()
    
    # outputs the continent(ufo and world wonder data)    
    elif ufo and wonders:  
        
        fig,ax = plt.subplots(figsize=(10,10))
        #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        
        continent_ufo_df = gpd.sjoin(geo_ufo,world_bound.loc[world_bound['continent'] == continent])
        continent_wonder_df = gpd.sjoin(geo_wonder,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        
        
        
        
        continent_ufo_df.plot(ax=ax,color='lime',markersize=15,marker='v',label='ufo_sightings')
        continent_wonder_df.plot(ax=ax,color='gold',markersize=15,marker='s',label='world_wonder')
        
        
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of all phenomena in the continent
        phenomenon_count = len(continent_ufo_df) + len(continent_wonder_df)
        phenomenon_percent = phenomena_percent(continent)['total_phenomenon_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12) 
        
        
        fig.show()
        
    # outputs the continent( meteorite and wonder data)       
    elif meteorite and wonders:  
        
        fig,ax = plt.subplots(figsize=(10,10))
        #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        continent_meteorites_df = gpd.sjoin(geo_meteorites,world_bound.loc[world_bound['continent'] == continent])
        continent_wonder_df = gpd.sjoin(geo_wonder,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        
        
        
        continent_meteorites_df.plot(ax=ax,color='orangered',markersize=15,label='metorites')
        continent_wonder_df.plot(ax=ax,color='gold',markersize=15,marker='s',label='world_wonder')
        
        
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of all phenomena in the continent
        phenomenon_count = len(continent_meteorites_df) + len(continent_wonder_df)
        phenomenon_percent = phenomena_percent(continent)['total_phenomenon_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {phenomenon_count} phenomena.
                    Which makes up {phenomenon_percent}% of all phenomena found across all continents.
                    """,color='w',ha='center',fontsize=12) 
        
        
        fig.show()
                
    # outputs the continent(meteorite data)
    elif meteorite:
        
        
        fig,ax = plt.subplots(figsize=(10,10))
         #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        continent_meteorites_df = gpd.sjoin(geo_meteorites,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        continent_meteorites_df.plot(ax=ax,color='orangered',markersize=15,label='metorites')
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of meteorites in the continent
        meteorite_count = len(continent_meteorites_df)
        meteorite_percent = phenomena_percent(continent)['meteorite_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {meteorite_count} meteorites.
                    Which makes up {meteorite_percent}% of all meteorites found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    {continent} has a total of {meteorite_count} meteorites.
                    Which makes up {meteorite_percent}% of all meteorites found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {meteorite_count} meteorites.
                    Which makes up {meteorite_percent}% of all meteorites found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {meteorite_count} meteorites.
                    Which makes up {meteorite_percent}% of all meteorites found across all continents.
                    """,color='w',ha='center',fontsize=12)
        
        
        fig.show()
        
    # outputs the continent(ufo data)
    elif ufo:
        
        
        fig,ax = plt.subplots(figsize=(10,10))
         #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        
        continent_ufo_df = gpd.sjoin(geo_ufo,world_bound.loc[world_bound['continent'] == continent])

        
        # geoplot the data
        continent_ufo_df.plot(ax=ax,color='lime',markersize=15,marker='v',label='ufo_sightings')
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of meteorites in the continent
        ufo_count = len(continent_ufo_df)
        ufo_percent = phenomena_percent(continent)['ufo_percent']
        
        ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    In {continent} there has been a total of {ufo_count} UFO sightings 
                    Which makes up {ufo_percent}% of all UFO sightings found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                    f"""
                    In {continent} there has been a total of {ufo_count} UFO sightings 
                    Which makes up {ufo_percent}% of all UFO sightings found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    In {continent} there has been a total of {ufo_count} UFO sightings 
                    Which makes up {ufo_percent}% of all UFO sightings found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    In {continent} there has been a total of {ufo_count} UFO sightings 
                    Which makes up {ufo_percent}% of all UFO sightings found across all continents.
                    """,color='w',ha='center',fontsize=12)
        
        
        fig.show()
        
    # outputs the continent(world wonders data)
    elif wonders:
        
        
        fig,ax = plt.subplots(figsize=(10,10))
        #locate the given continent and plot it
        # locate the given continent and plot it
        # The intial plot of Oceania and Europe had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for these continents
        
        if continent == 'Oceania':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into oceania
            plt.axis(oceania_proj)
            
            
        elif continent == 'Europe':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            # zoom into europe
            plt.axis(eu_proj)
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
        
        
        # locate the data points that fall in the continent with a spatial join
        
        continent_wonder_df = gpd.sjoin(geo_wonder,world_bound.loc[world_bound['continent'] == continent])
        
        # geoplot the data
        continent_wonder_df.plot(ax=ax,color='gold',markersize=15,marker='s',label='world_wonder')
        
        #remove the y and x axis labels
        plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
        # plot interpretability
        ax.set_facecolor('k')
        fig.set_facecolor('dimgray')
        plt.title("Phenomenon of " + str(continent),fontsize=20,color='w')
        plt.legend(markerscale=1,title='Phenomenon')
        ## plot caption
        ### accquire the count and percent of meteorites in the continent
        wonder_count = len(continent_wonder_df)
        wonder_percent = phenomena_percent(continent)['wonder_percent']
        
       ### adjust the caption position on the continent that is passed in
        if continent == 'Oceania':
            
            plt.figtext(0.49,0.10,
                    f"""
                    {continent} has a total of {wonder_count} wonders.
                    Which makes up {wonder_percent}% of all wonders found across all continents.
                    """,color='w',ha='center',fontsize=12)
        elif (continent == 'Africa') | (continent == 'South America'):
            plt.figtext(0.49,0.05,
                   f"""
                    {continent} has a total of {wonder_count} wonders.
                    Which makes up {wonder_percent}% of all wonders found across all continents.
                    """,color='w',ha='center',fontsize=12)
            
        elif continent == 'Antarctica':
            plt.figtext(0.49,0.40,
                    f"""
                    {continent} has a total of {wonder_count} wonders.
                    Which makes up {wonder_percent}% of all wonders found across all continents.
                    """,color='w',ha='center',fontsize=12)
            plt.legend(markerscale=1,title='Phenomenon',bbox_to_anchor=(0.15,0.10))
        
         
        else:
            plt.figtext(0.49,0.2,
                    f"""
                    {continent} has a total of {wonder_count} wonders.
                    Which makes up {wonder_percent}% of all wonders found across all continents.
                    """,color='w',ha='center',fontsize=12)
        
        
        fig.show()
        
    # outputs the continent(no phenomenon)
    else:
        
        fig,ax = plt.subplots(figsize=(10,10))
        # locate the given continent and plot it
        # The intial plot of Oceania had an irregular xmin value. In retrun, the plot had excess space
        # this conditional configures the plot for Oceania
        
        if continent == 'Oceania':
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            #remove the y and x axis labels
            plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
            ax.set_facecolor('k')
            fig.set_facecolor('dimgray')
            plt.title(continent,fontsize=20,color='w')
           # zoom into oceania
            plt.axis(oceania_proj)
            
            plt.show()
            
        elif continent == 'Europe':
            
            world_bound.loc[world_bound['continent'] == continent].plot(ax=ax,edgecolor='w',color='k')
            #remove the y and x axis labels
            plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
            ax.set_facecolor('k')
            fig.set_facecolor('dimgray')
            plt.title(continent,fontsize=20,color='w')
            # zoom into europe
            plt.axis(eu_proj)
            
            plt.show()
            
        else:
            continent_df = world_bound.loc[world_bound['continent'] == continent]
            continent_df.plot(ax=ax,edgecolor='w',color='k')
            
            #remove the y and x axis labels
            plt.tick_params(top=False, bottom=False, left=False, right=False,
                labelleft=False, labelbottom=False)
            ax.set_facecolor('k')
            fig.set_facecolor('dimgray')
            plt.title(continent,fontsize=20,color='w')
            

            plt.show()
        

```
